```python
import os
import pathlib
import numpy as np
import librosa
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data_dir = tf.keras.utils.get_file(
    'mini_speech_commands.zip',
    origin="https://storage.googleapis.com/download.tensorflow.org/data/mini_speech_commands.zip",
    extract=True
)
data_dir = pathlib.Path(data_dir).parent / 'mini_speech_commands'

commands = np.array(tf.io.gfile.listdir(str(data_dir)))
commands = commands[commands != 'README.md']
print("Commands:", commands)
```

```python
def add_noise(audio):
    noise = np.random.normal(0, 0.005, audio.shape)
    return audio + noise

def preprocess_audio(file_path, max_len=124):
    audio, sr = librosa.load(file_path, sr=16000)
    spectrogram = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=128)
    log_spectrogram = librosa.power_to_db(spectrogram)
    audio = add_noise(audio) 
    
    # Pad or truncate to fixed length
    if log_spectrogram.shape[1] < max_len:
        pad_width = max_len - log_spectrogram.shape[1]
        log_spectrogram = np.pad(log_spectrogram, ((0, 0), (0, pad_width)), mode='constant')
    else:
        log_spectrogram = log_spectrogram[:, :max_len]

    return log_spectrogram


   
```

```python
filenames = tf.io.gfile.glob(str(data_dir) + '/*/*.wav')
labels = [pathlib.Path(f).parent.name for f in filenames]

label_to_index = {label: i for i, label in enumerate(commands)}
y = np.array([label_to_index[label] for label in labels])

X_train, X_val, y_train, y_val = train_test_split(filenames, y, test_size=0.2, random_state=42)

def files_to_dataset(file_list, label_list):
    spectrograms = [preprocess_audio(f) for f in file_list]
    spectrograms = np.array([np.expand_dims(s, axis=-1) for s in spectrograms])
    
    # Normalize to 0-1 range
    spectrograms = (spectrograms - spectrograms.min()) / (spectrograms.max() - spectrograms.min())
    
    return spectrograms, np.array(label_list)
```

```python
X_train_tensor, y_train_tensor = files_to_dataset(X_train, y_train)
X_val_tensor, y_val_tensor = files_to_dataset(X_val, y_val)

print("Train shape:", X_train_tensor.shape)
print("Validation shape:", X_val_tensor.shape)
```

```python
import sounddevice as sd
import soundfile as sf

def record_audio(duration=1, fs=16000, output='test.wav'):
    print("Recording... 🎙️")
    audio = sd.rec(int(duration * fs), samplerate=fs, channels=1, dtype='float32')
    sd.wait()  # Wait until recording is finished
    sf.write(output, audio, fs)
    print("Recording saved as", output)
```

```python
def preprocess_recorded_audio(filename, target_shape=(128, 124)):
    audio, sr = librosa.load(filename, sr=16000)
    spectrogram = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=target_shape[0])
    log_spectrogram = librosa.power_to_db(spectrogram)
    audio, _ = librosa.effects.trim(audio)  # Remove silence at start/end
    audio = audio / np.max(np.abs(audio))  # Normalize waveform amplitude

    # Normalize
    log_spectrogram = (log_spectrogram - log_spectrogram.min()) / (log_spectrogram.max() - log_spectrogram.min())
    
    # Pad or trim to fixed width
    if log_spectrogram.shape[1] < target_shape[1]:
        pad_width = target_shape[1] - log_spectrogram.shape[1]
        log_spectrogram = np.pad(log_spectrogram, ((0, 0), (0, pad_width)), mode='constant')
    else:
        log_spectrogram = log_spectrogram[:, :target_shape[1]]
    
    return np.expand_dims(log_spectrogram, axis=-1)
```

```python
import numpy as np
import soundfile as sf

def generate_t3_alarm(fs=16000, duration=8, freq=520):
    t = np.linspace(0, duration, int(fs * duration), False)
    signal = np.zeros_like(t)

    cycle_length = 4  # 0.5s * 3 + 1.5s pause
    samples_per_cycle = int(cycle_length * fs)

    for i in range(0, len(t), samples_per_cycle):
        for j in range(3):
            start = i + int((j * 0.5) * fs)
            end = start + int(0.5 * fs)
            if end > len(t): break
            signal[start:end] = 0.5 * np.sin(2 * np.pi * freq * t[start:end])

    return signal

audio = generate_t3_alarm()
sf.write("fire_t3.wav", audio, 16000)
```

```python
import numpy as np
import soundfile as sf
import os

def generate_t3_alarm(fs=16000, duration=6, freq=520, noise=False):
    t = np.linspace(0, duration, int(fs * duration), False)
    signal = np.zeros_like(t)
    cycle_length = 4  # total cycle time (3 beeps + pause)

    for i in range(0, len(t), int(cycle_length * fs)):
        for j in range(3):
            start = i + int((j * 0.5) * fs)
            end = start + int(0.5 * fs)
            if end >= len(t): break
            signal[start:end] = 0.5 * np.sin(2 * np.pi * freq * t[start:end])

    # Optional: add some background noise
    if noise:
        signal += np.random.normal(0, 0.01, len(signal))
    
    return signal

# Generate 200 positive samples
os.makedirs("dataset/t3", exist_ok=True)
for i in range(200):
    f = 520 + np.random.randint(-50, 50)  # vary ±10%
    s = generate_t3_alarm(freq=f, noise=True)
    sf.write(f"dataset/t3/t3_{i}.wav", s, 16000)
```

```python
def generate_non_t3(fs=16000, duration=6, freq=None):
    t = np.linspace(0, duration, int(fs * duration), False)
    if freq:
        signal = 0.5 * np.sin(2 * np.pi * freq * t)
    else:
        signal = np.random.normal(0, 0.01, len(t))  # white noise
    return signal

os.makedirs("dataset/not_t3", exist_ok=True)
for i in range(200):
    choice = np.random.choice(['tone', 'noise'])
    if choice == 'tone':
        f = np.random.choice([440, 880, 1000, 250])  # non-T3 tones
        s = generate_non_t3(freq=f)
    else:
        s = generate_non_t3()
    sf.write(f"dataset/not_t3/not_t3_{i}.wav", s, 16000)
```
